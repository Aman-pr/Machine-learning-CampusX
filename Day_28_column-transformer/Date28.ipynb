#Column Transformer

```python
import pandas as pd
```

```python
df=pd.read_csv("covid_toy.csv")
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
transformer=ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(drop='first'),['gender','city'])
    ],remainder='passthrough')
```

```python
convo_train=transformer.fit_transform(X_train)
```

```python
conv_test=transformer.fit_transform(X_train)
```

```python
X_train_new = pd.DataFrame(convo_train[:, :X_train.shape[1]], columns=X_train.columns)
X_test_new = pd.DataFrame(convo_test[:, :X_test.shape[1]], columns=X_train.columns)
```
